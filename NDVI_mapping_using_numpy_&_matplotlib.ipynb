```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming ndvi_values is a NumPy array containing your NDVI values ranging from -1 to 1
# You might need to adjust the range based on your specific NDVI values
ndvi_values = np.random.uniform(-1, 1, size=(100, 100))  # Example random NDVI values

# Define a color map for NDVI
cmap = plt.cm.RdYlGn  # Red-Yellow-Green color map is commonly used for NDVI

# Normalize NDVI values to the range [0, 1] for proper colormap mapping
norm_ndvi = (ndvi_values - (-1)) / (1 - (-1))

# Create a plot with the NDVI values and apply the color map
plt.imshow(norm_ndvi, cmap=cmap)
plt.colorbar(label='NDVI')
plt.title('NDVI Visualization')
plt.show()
```
